```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
```

# Subplots
- In One Figure Can Draw More Than One Plot

```python
dev_x = [30, 31, 32, 33, 34, 35, 36]
dev_y = [10365, 15364, 20236, 20478, 25698, 30785, 35156]

py_dev_x = [30, 31, 32, 33, 34, 35, 36]
py_dev_y = [32154, 33659, 35478, 38915, 42159, 45987, 48368]

fig, ax = plt.subplots()

# first figure 
ax.plot(dev_x, dev_y, color="green", marker="o", label="Mean Developer Salary")

# second figure
ax.plot(py_dev_x, py_dev_y, color="red", marker="o", label="Mean Python Salary")

ax.set_title("Sala For Each Age")
ax.set_xlabel("Age")
ax.set_ylabel("Salary")
ax.legend()
plt.grid()

plt.show()

# save the figure 
fig.savefig("AgeSalary")
```

# Show Each Plot In One Figure

```python
# Sample data
py_dev_x = [25, 30, 35, 40, 45, 50]
py_dev_y = [40000, 50000, 60000, 55000, 70000, 65000]

# Create a figure and a set of subplots
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# Plotting in the second subplot
ax[1].set_title("Salary For Each Python Developer")
ax[1].set_xlabel("Age")
ax[1].set_ylabel("Salary")
ax[1].plot(py_dev_x, py_dev_y, color="red", marker="o", label="Mean Python Developer Salary", linewidth=2)
ax[1].legend()

# ======================================
# == Adjust layout to prevent overlap == 
# ======================================
plt.tight_layout()

##########################################################################################################

dev_x = [30, 31, 32, 33, 34, 35, 36]
dev_y = [10365, 15364, 20236, 20478, 25698, 30785, 35156]

# Plotting in the second subplot
ax[0].set_title("Salary For Each Developer")
ax[0].set_xlabel("Age")
ax[0].set_ylabel("Salary")
ax[0].plot(dev_x, dev_y, color="blue", marker="o", label="Mean Developer Salary", linewidth=2)
ax[0].legend()

# Show the plot
plt.show()
```

# Share The Axis
- Can Share Y-axis By Using "sharey"
- Can Share X-axis By Using "sharex"
## Change The Size Of The Figure
- By Using figsize(width, height), can resize the figure

```python
# Sample data
py_dev_x = [25, 30, 35, 40, 45, 50]
py_dev_y = [40000, 50000, 60000, 55000, 70000, 65000]

# Create a figure and a set of subplots
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5), sharey=True)

# Plotting in the second subplot
ax[1].set_title("Salary For Each Python Developer")
ax[1].set_xlabel("Age")
ax[1].set_ylabel("Salary")
ax[1].plot(py_dev_x, py_dev_y, color="red", marker="o", label="Mean Python Developer Salary", linewidth=2)
ax[1].legend()

# ======================================
# == Adjust layout to prevent overlap == 
# ======================================
plt.tight_layout()

##########################################################################################################

dev_x = [30, 31, 32, 33, 34, 35, 36]
dev_y = [10365, 15364, 20236, 20478, 25698, 30785, 35156]

# Plotting in the second subplot
ax[0].set_title("Salary For Each Developer")
ax[0].set_xlabel("Age")
ax[0].set_ylabel("Salary")
ax[0].plot(dev_x, dev_y, color="blue", marker="o", label="Mean Developer Salary", linewidth=2)
ax[0].legend()

# Show the plot
plt.show()
```
