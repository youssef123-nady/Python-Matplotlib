```python
import numpy as np
import matplotlib.pyplot as plt 

# Use The Following Command To Make The Chart Inline The Notebook
%matplotlib inline

# Change The Style Of The Graph
plt.style.use("ggplot")
```

# Pie Charts

```python
data = [50, 20, 15, 30]
labels = ["Python", "Js", "PHP", "MySQL"]

# Make Pie Chart Graph
plt.pie(data, labels=labels)

plt.title("A Pie Chart")
plt.show()
```

# Change The Color Of Edges

```python
data = [50, 20, 15, 30]
labels = ["Python", "Js", "PHP", "MySQL"]

# Change The Color Of Edge 
# Make Pie Chart Graph
plt.pie(data, labels=labels, wedgeprops={"edgecolor": "black"})

plt.title("A Pie Chart")
plt.show()
```

# Change The Color Of Each Elemenet

```python
data = [50, 20, 15, 30]
labels = ["Python", "Js", "PHP", "MySQL"]

# Change The Color Of Each Element By Using Hexadecimal values
colors = ["orange", "red", "green", 'blue']
plt.pie(data, labels=labels, colors=colors, wedgeprops={"edgecolor": "#000"})

plt.title("A Pie Chart")
plt.show()
```

# Change The Color Of Each Elemenet Using Hexadeciamal Values

```python
data = [50, 20, 15, 30]
labels = ["Python", "Js", "PHP", "MySQL"]

# Change The Color Of Each  By Using Hex Values
colors = ["#EC01F7", "#F70114", "#48DD99", '#F7E801']
plt.pie(data, labels=labels, colors=colors)

plt.title("A Pie Chart")
plt.show()
```

# Explode An ELement
- maybe you want one of the wedges to stand out? the explode parameter allows you to do that

```python
data = [50, 20, 15, 30]
labels = ["Python", "Js", "PHP", "MySQL"]

s = [0, 0.1, 0, 0]
plt.pie(data, labels=labels, explode=s)

plt.title("A Pie Chart")
plt.show()
```

# Make Shadow To The Pie Chart
- make shadow to wedges by using the attribute "shadow=True"

```python
data = [50, 20, 15, 30]
labels = ["Python", "Js", "PHP", "MySQL"]

x = [0.1, 0, 0, 0]
plt.pie(data, labels=labels, explode=x, shadow=True)

plt.title("A Pie Chart")
plt.show()
```

# Determine The Angle That The Chart Will Start From 

```python
data = [50, 20, 15, 30]
labels = ["Python", "Js", "PHP", "MySQL"]

x = [0.05, 0, 0, 0]
plt.pie(data, labels=labels, explode=x, startangle=45)

plt.title("A Pie Chart")
plt.show()
```

# Write The Percentage Number To Wedges
- 0.1f: refers to one point number on the floating number
- 0.2f: referes to two point numbers on the floating number, and so on 

```python
data = [50, 20, 15, 30]
labels = ["Python", "Js", "PHP", "MySQL"]

x = [0.1, 0, 0, 0]
plt.pie(data, labels=labels, explode=x, autopct="%0.2f%%")

plt.title("A Pie Chart")
plt.show()
```

# Rotate The Title Of Each Wedges

```python
data = [50, 20, 15, 30]
labels = ["Python", "Js", "PHP", "MySQL"]

x = [0.1, 0, 0, 0]
plt.pie(data, labels=labels, explode=x, autopct="%0.2f%%", rotatelabels=True)

plt.title("A Pie Chart")
plt.show()
```
